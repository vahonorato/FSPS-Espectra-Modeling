```python
import fsps
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u 

plt.style.use('bmh')
```

```python
#galaxia espiral
sp = fsps.StellarPopulation(zcontinuous=1)
sp.params["imf_type"] = 2 # Salpeter IMF (0)
sp.params["sfh"] = 1  # Tau model SFH (1)
sp.params["dust_type"] = 2
sp.params["logzsol"] = 0
sp.params["dust2"] = 0.2

tau = [0.05, 3, 7, 13.7] 
colors = ['darkred', 'darkblue', 'darkgreen', 'darkorange', 'purple', 'magenta']

for t,i in enumerate(tau):
    sp.params['tau'] = i
    wave, spec = sp.get_spectrum(tage = 13.7)
    plt.plot(wave, spec, lw = 1.5, alpha = 1, color = colors[t], label = r'$\tau$ = '+str(i))

plt.yscale('log')
plt.xscale('log')
plt.xlim(500, 10**8)
plt.ylim(10**-21, 10**-12.8)
plt.xlabel(r'$\lambda$ [$\AA$]')
plt.ylabel(r'L${\nu}$ [L$_{\odot}$ Hz$^{-1}$]')
plt.grid(lw = 1)
plt.axvspan(1000, 4000, color='black', alpha=0.6, ls=  '--', lw = 1, label = r'UV')
plt.axvspan(4000, 7500, color='dimgray', alpha=0.7, ls=  '--', lw = 1, label = r'Optical')
plt.axvspan(7500, 10**7, color='darkgray', alpha=0.42, ls=  '--',  lw = 1, label = r'IR')

plt.legend(fontsize = 11, framealpha = 0.3, loc = 'lower center', edgecolor = (0,0,0,1))

plt.tight_layout()
plt.savefig(f'Images/different_efoldingtime_spiral.png', dpi = 600)

plt.show()
```
